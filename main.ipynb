```python
import pandas as pd
```

```python
movies = movies.dropna(subset=['genres'])
```

```python
null_keywords = movies[movies["keywords"].isna() == True]
null_keywords["keywords"] = null_keywords["genres"]
null_keywords
```

```python
movies = pd.merge(movies, null_keywords, how = 'outer')
movies
```

```python
movies = movies.dropna(subset=['keywords'])
movies
```

```python
num_vals = ["budget", "runtime", "budget_adj"]
for col in num_vals:
    movies[col].replace(0, movies[col].mean(), inplace=True)
```

```python
movies
```

```python
def replace_char(text):
    if "|" in text:
        text = text.lower()
        text = text.replace("|", " ")
    return text

movies["genres"] = movies["genres"].apply(replace_char)
movies["keywords"] = movies["keywords"].apply(replace_char)
movies["tags"] = movies["genres"] + movies["keywords"]
movies_cpy = movies
movies = movies.drop(columns = ["cast","homepage","director","cast", "release_date", "revenue_adj","imdb_id","revenue","tagline","overview", "production_companies", "original_title", "release_year"])
movies_cpy = movies_cpy[["id", "original_title", "tags"]]
movies = movies.drop(columns =["keywords", "genres"])
```

```python
movies
```

```python
movies_cpy
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tf = TfidfVectorizer()
vector = tf.fit_transform(movies["tags"].values.astype('U')).toarray()
vector
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vector)
similarity
```

```python
distance = sorted(list(enumerate(similarity[2])), reverse= True, key = lambda vector : vector[1])
```

```python
for i in distance[0:5]:
    print(movies_cpy.iloc[i[0]].original_title)
```

```python
def recommand(movies):
    index = movies_cpy[movies_cpy["original_title"] == movies].index[0]
    distance = sorted(list(enumerate(similarity[index])), reverse= True, key = lambda vector : vector[1])
    for i in distance[0:5]:
        print(movies_cpy.iloc[i[0]].original_title)
```

```python
recommand("Iron Man 3")
```
